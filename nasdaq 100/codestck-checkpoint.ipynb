```python
import pandas as pd
import numpy as np
from datascience import *
import os
import glob
```

```python
stocks_return = pd.DataFrame(columns = ['Ticker','Year', 'Annual Return'])
test = pd.DataFrame(columns = ['Ticker','Year', 'Annual Return'])
total_stocks = []
```

```python
aapl = pd.read_csv("AAPL.csv")
```

```python
def conv(data):
    data['Date'] = pd.to_datetime(data['Date']) #Convert data type of 'Date' to datetime
```

```python
def monthlyReturn(df, date):
    mask = (df['Date'].dt.to_period('M') == pd.to_datetime(date).to_period('M')) 
    t = df.loc[mask] #Returns daily return of the month
    t = t.iloc[[0,-1]] #Select the first day and last day of the month
    Open = t.iloc[0,1] #Extract the Open value of 1st row
    Close = t.iloc[1,5] #Extract the AdjClose value of 2nd row
    return (Close - Open)/Open * 100 #Returns percentage return of the month
```

```python
def yearlyReturn(df, date):
    mask = (df['Date'].dt.to_period('Y') == pd.to_datetime(date).to_period('Y')) 
    t = df.loc[mask] #Returns daily return of the year
    t = t.iloc[[0,-1]]#Select the first day and last day of the year
    Open = t.iloc[0,1] #Extract the Open value of 1st row
    Close = t.iloc[1,5] #Extract the AdjClose value of 2nd row
    return (Close - Open)/Open * 100 #Returns percentage return of the year
```

```python
def getAnnualReturn(stock, name, data, startYear, endYear):
    temp_year = []
    temp_return = []
    for i in range(startYear, endYear):
        temp_year.append(i)
        temp_return.append(yearlyReturn(stock, str(i)))
    data['Year'] = temp_year
    data['Annual Return'] = temp_return
    data['Ticker'] = name
```

```python
files = glob.glob()
for i in files:
    temp = pd.read_csv(i)
    conv(temp)
    tp = pd.DataFrame(columns = ["Ticker",'Year','Annual Return'])
    try:
        print("Now testing " + i)
        getAnnualReturn(temp, str(os.path.basename(i)), tp, max(2000, temp['Date'].dt.year[0]),2021)
        test = test.append([tp])
    except:
        print("Ignoring " + i)
        continue
```

```python
#pd.options.display.max_rows = None
#test.loc[test["Ticker"]=="CCAP.csv"]
data1 = Table.from_df(test)
data1.show()
```

```python
data2 = data1.group('Ticker',np.average).sort('Annual Return average', descending = True)
data2.show()
```
